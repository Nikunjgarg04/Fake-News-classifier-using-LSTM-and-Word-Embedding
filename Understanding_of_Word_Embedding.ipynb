```
from tensorflow.keras.preprocessing.text import one_hot
```

```
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```
###Vocablary size is basically the dictionary size
voc_size=10000
```

One Hot Representation

```
###One hot encoding you will be getting the indexes from the dictionary
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

**Word Embedding Representation**

```
##to make a embedding matrix
###We are using pad_sequences because we want the same number of words in all the sentences. It is very important
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential 
```

```
import numpy as np
```

```
###Here I need to give the sentence length, to which number I want to have the sentence length. Padding is of two types pre and post.
### In pre padding it will add the number in the beginning of the sentence to make the length of the sentence equals to the given sentence length.
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```
###dimension is basically the number of features
dim=10
```

```
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```
model.summary()
```

```
print(model.predict(embedded_docs))
```
